```python
import numpy as np
import sympy as sp
import numpy.linalg as LA
from matplotlib.pyplot import subplots as spp
```

```python
def bdg_cos(k):
    # Bogoliubov transformation
    sigma_3 = np.diag([1, 1, -1, -1])

    H = 1/2*np.array([[1,0,0,np.cos(k)],[0,1,np.cos(k),0],[0,np.cos(k),1,0],[np.cos(k),0,0,1]])

    H_BdG = np.matmul(sigma_3, H)
    eigval_k,eigvec_k = (LA.eig(H_BdG))

    idx = np.real(eigval_k).argsort()[::-1]
    eigval_k = np.real(eigval_k[idx])
    eigvec_k = eigvec_k[:, idx]
    return eigval_k, eigvec_k
```

```python
def bdg_exp(k):
    # Bogoliubov transformation
    sigma_3 = np.diag([1, 1, -1, -1])

    H = 1/2*np.array([[1,0,0,np.exp(1j*k)],[0,1,np.exp(-1j*k),0],[0,np.exp(1j*k),1,0],[np.exp(-1j*k),0,0,1]])

    H_BdG = np.matmul(sigma_3, H)
    eigval_k,eigvec_k = (LA.eig(H_BdG))

    idx = np.real(eigval_k).argsort()[::-1]
    eigval_k = np.real(eigval_k[idx])
    eigvec_k = eigvec_k[:, idx]
    return eigval_k, eigvec_k
```

```python
def bdg_2cos(k):
    # Bogoliubov transformation
    sigma_3 = np.diag([1, 1, -1, -1])

    H = 1/2*np.array([[1,0,0,2*np.cos(k)],[0,1,0,0],[0,0,1,0],[2*np.cos(-k),0,0,1]])

    H_BdG = np.matmul(sigma_3, H)
    eigval_k,eigvec_k = (LA.eig(H_BdG))

    idx = np.real(eigval_k).argsort()[::-1]
    eigval_k = np.real(eigval_k[idx])
    eigvec_k = eigvec_k[:, idx]
    return eigval_k, eigvec_k
```

```python
i = 0
n=100
dk=2*np.pi/n

eng_all = np.empty((0,4), float)
fig, ax = spp(facecolor='0.95')

while i < n+1:
    eng, vec = bdg_cos(-np.pi+dk*i)
    eng_all = np.append(eng_all, [eng], axis=0)
    print(eng)
    i += 1


num = int(eng_all.size/4)
x = np.linspace(-np.pi, np.pi, num, retstep=False, dtype=None, axis=0)

fig.gca().set_title("Band structure for 1D AFM Heisenberg model")
fig.gca().set_xlabel("k")
fig.gca().set_ylabel("$Îµ_{nk}\ meV$")

eng_all = np.reshape(eng_all,(-1,4))
fig.gca().plot(x, eng_all[:,0], color='k')
fig.gca().plot(x, eng_all[:, 1], color='k')
#fig.gca().plot(x, eng_all[:, 2], color='k')
#fig.gca().plot(x, eng_all[:, 3], color='k')

fig.savefig('band_cos.png')
```
